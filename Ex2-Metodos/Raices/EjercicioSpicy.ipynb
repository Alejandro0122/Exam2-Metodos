```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
def f(x):
    return (x**5 + 5.5 * x**4 + 4.5 * x**3 - 9.5 * x**2 - 15.5 * x - 6)
def f_prime(x):
    return (5 * x**4 + 22 * x**3 + 13.5 * x**2 - 19 * x - 15.5)
def newton_raphson(x0, tol=1e-5, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        if fpx == 0:
            raise ValueError("Derivative is zero. No solution found.")
        x_new = x - fx / fpx
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    raise ValueError("Exceeded maximum iterations. No solution found.")

for x0 in np.linspace(-5, 5, 100):
    try:
        root = newton_raphson(x0)
        print(f"Root found: {root}")
    except ValueError as e:
        print(f"Error for initial guess {x0}: {e}")
roots = []

x0 = [] 
for x0 in x0:
    try:
        root = newton_raphson(x0)
        roots.append(root)
        print(f"Root found: {root}")
    except ValueError as e:
        print(f"Error for initial guess {x0}: {e}")


f = lambda x, a: x**5 + 5.5 * x**4 + 4.5 * x**3 - 9.5 * x**2 - 15.5 * x - 6 
fder = lambda x, a: (5 * x**4 + 22 * x**3 + 13.5 * x**2 - 19 * x - 15.5)
rng = np.random.default_rng()
x = rng.standard_normal(100)
a = np.arange(-50, 50)
vec_res = optimize.newton(f, x, fprime=fder, args=(a, ), maxiter=200)
plt.plot(a, vec_res)
plt.show()
```
